The focus of this project will be a Pokémon battle game in which we made special Pokémon objects with specific advantages and traits. The names, types, moves, and unique abilities related to attack, defense, and general health of each Pokémon serve to characterize them. The game mimics these Pokémon engaging in battle on turns while taking type advantages and move effectiveness into account.





```python



# Delay printing
import sys
import time
import numpy as np

def delay_print(s):
    # print one character at a time
    for c in s:
        sys.stdout.write(c)
        sys.stdout.flush()
        time.sleep(.1)

# Create the class
class Pokemon:
    def __init__(self, name, types, moves, EVs, health='==================='):
        # save variables as attributes
        self.name = name
        self.types = types
        self.moves = moves
        self.attack = EVs['ATTACK']
        self.defense = EVs['DEFENSE']
        self.health = health
        self.bars = 20  # Amount of health bars

    def fight(self, opponent):
        # Simulate a turn-based battle between two Pokemon

        delay_print(f"\n\n-----{self.name} is battling {opponent.name}-----\n")

        # Continue battling until one of the Pokemon faints
        while self.bars > 0 and opponent.bars > 0:
            # Pokemon take turns attacking
            self.attack_opponent(opponent)
            if opponent.bars <= 0:
                break  # Opponent fainted

            opponent.attack_opponent(self)
            if self.bars <= 0:
                break  # Self fainted

        # Determine the winner
        if self.bars <= 0:
            delay_print(f"\n\n{self.name} has fainted. {opponent.name} is the winner!\n")
        else:
            delay_print(f"\n\n{opponent.name} has fainted. {self.name} is the winner!\n")

    def attack_opponent(self, opponent):
        # Simulate an attack on the opponent

        # Select a random move from the available moves
        move = self.moves[0]  # For simplicity, using the first move in the list

        delay_print(f"\n{self.name} used {move}!")

        # Check if the move is super effective or not very effective based on types
        if self.types == opponent.types:
            delay_print("\nIt's not very effective...")
        elif opponent.types in ['Dark', 'Psychic', 'Fighting']:
            delay_print("\nIt's super effective!")

        # Simulate damage calculation (you can customize this based on your requirements)
        damage = self.attack - opponent.defense
        opponent.bars -= damage  # Reduce opponent's health

        # Add the back bars plus defense and boost
        opponent.health = "=" * int(opponent.bars + 0.1 * opponent.defense)

        delay_print(f"\n{opponent.name}'s health: {opponent.health}")
        
    

# Example Usage
if __name__ == '__main__':
    # Create Pokemon objects
    pokemon1 = Pokemon("Charizard", "Fire", ["Ember", "Flamethrower"], {"ATTACK": 84, "DEFENSE": 78})
    pokemon2 = Pokemon("Blastoise", "Water", ["Water Gun", "Hydro Pump"], {"ATTACK": 83, "DEFENSE": 100})
    pokemon3 = Pokemon("Alakazam", "Psychic", ["Confusion", "Psychic"], {"ATTACK": 50, "DEFENSE": 45})
    pokemon4 = Pokemon("Machamp", "Fighting", ["Karate Chop", "Cross Chop"], {"ATTACK": 130, "DEFENSE": 80})

    # Create teams
    #team1 = [pokemon1, pokemon2]
    #team2 = [pokemon3, pokemon4]

    # Simulate battles between corresponding Pokemon in team1 and team2
    for i in range(2):
        pokemon4.fight(pokemon3)
```
